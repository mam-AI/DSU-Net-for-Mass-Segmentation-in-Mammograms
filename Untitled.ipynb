```python

import numpy as np
import matplotlib.pyplot as plt
import pydicom as dcm 
import cv2 as cv
```

```python
example_file = "ejemplo/manJRO3m7Z/CC_R_manJRO3m7Z_30de85bb6af85994.dcm" #Imagen craneocaudal derecha ya anonimizada ;)
ds = dcm.dcmread(example_file)

plt.imshow(ds.pixel_array,cmap=plt.cm.gray_r)
```

Nice, podemos ver la imagen sin mayor problema. Ahora a ver su histograma y otra informacion relevante

```python
img = ds.pixel_array
```

```python
hist, bins = np.histogram(img.flatten(), 256, [img.min(), img.max()])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()

fig, ax = plt.subplots(ncols=2, figsize=(12, 8))
ax[0].imshow(img, cmap=plt.cm.gray_r)
ax[0].axis('off')

#ax[1].plot(cdf_normalized, color='b')
ax[1].hist(img.flatten(), 256, [img.min(), img.max()], color='Orange')
ax[1].set_ylim([0,1e5])
```
